```python
def corr(x1, x2, theta):
    k = np.zeros((x1.shape[0],x2.shape[0]))
    for i in range(x1.shape[0]):
        k[i,:] = np.exp(-np.sum(theta*(x1[i,:]-x2)**2, axis=1))
    return k
```

```python
k = corr(df4Train.to_numpy(), df4Train.to_numpy(), theta)
```

```python
def negLogLikelihood(theta):
    
    theta = 10**theta
    n = df4Train.to_numpy().shape[0]
    one = np.ones((n,1))
    
    K = corr(df4Train.to_numpy(), df4Train.to_numpy(), theta) + np.eye(n)*1e-9
    invK = np.linalg.inv(K)
    
    mean = (one.T@invK@yTrain.to_numpy())/(one.T@invK@one)
    
    var = (yTrain.to_numpy()-mean*one).T@invK@(yTrain.to_numpy()-mean*one)/n
    #1/n.(y-1.u).T.invK.(y-1.u)
    
    detK = np.linalg.det(K)
    
    lnLikeli = -(n/2)*np.log(var) - 0.5*np.log(detK)
    
    return lnLikeli.flatten()
```

Emsemble Model with Linear Regression as Meta Learner

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
```

```python
def getStacking():
    level0 = []
    level0.append(('XGR', XGBRegressor()))
    level0.append(('SVR', SVR()))
    
    level1 = LinearRegression()
    
    model = StackingRegressor(estimators=level0, final_estimator= level1, cv=5)
    
    return model
```

```python
def evaluate_model(model, X, y):
    cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
    scores = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv)
    return scores
```
